Bar Plot

```python
import matplotlib.pyplot as plt
x = [98,78,45,65,45]
y = ['part1','part2','part3','part4','part5']
colour = ['red','pink','blue','skyblue','yellow']
plt.bar(y,x,color=colour)
plt.xlabel('popularity')
plt.ylabel("Part's of Harry potter")
plt.title("Popularity of different parts of harry potter")
plt.show()
```

Pie Chart


```python
import matplotlib.pyplot as plt
brand = ["Oneplus", "Apple", "samsumng","Nokia", "Redmi"]
x = [22,35,30,13,10]
c = ["red","silver","blue","pink","orange"]
#Eplored
ex = [0,0,0,0,0.2]
plt.pie(x, labels= brand,colors=c, explode=ex,shadow="True", autopct="%.2f",startangle=180)
#Saving the image in your device
plt.savefig("pie_chart.png",facecolor = "green", pad_inches = 0.4, bbox_inches = "tight")
plt.show()
```

Legends

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [45,67,33,67,12]
y1 = [41,60,13,66,13]
y2 = [11,15,23,21,30]
y3 = [30,25,41,36,40]

plt.figure(figsize = [5,3])
plt.plot(x,y, label = "male")
plt.plot(x,y1, label ="female")
plt.plot(x,y2, label = "kids")
plt.plot(x, y3, label = "blank")
plt.legend(bbox_to_anchor = (0.8,1.2), ncols = 2, labelspacing = 3)
plt.show()
```
